# Parts of EDA 
1. Univariate Analysis ==> Analysis on a single Independent column 
2. Bivariate Analysis ==> Analysis on two columns 
3. Multivariate Analysis ==> Analysis on more than 2 columns 

```python
# Data Types 
# 1. Numerical Data 
# 2. Categorical Data 
```

```python
import numpy as np 
import pandas as pd 
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("C:\\Users\\saurabh\\Desktop\\pyth\\titanic.csv")
```

```python
df.head() 
```

# 1. Univariate Analysis 

```python
df.columns
```

```python
sns.countplot(x = df['Survived'])
```

```python
# If we want to see acutal values then we use value_counts() . 

df['Survived'].value_counts() 
```

```python
df['Survived'].value_counts().plot(kind = 'bar') 
```

```python
# If we want to check for Pclass then 

sns.countplot(df['Pclass'])
```

```python
df['Pclass'].value_counts() 
```

```python
# If we want to find out percentage then use piechart . 
df['Survived'].value_counts() 
```

```python
df['Survived'].value_counts().plot(kind = 'pie' , autopct = '%.2f') 
```

```python
# If We have Numerical data then we use Histogram
# because it finds the distribution . 
```

```python
plt.hist(df['Age']) 
```

```python
# 2 . Distplot 
# curve ==> KDE(Kernal Density Extraction) 
# use for find Probabilty 
```

```python
sns.distplot(df['Age']) 
```

```python
sns.distplot(df['Age'] , hist = False) 
```

```python
# 3 . Boxplot 
# for find our outliers 

# 1 . lower fence
# 2 . 25% data
# 3.  IQR(Inter Quartile Range)(75%-25%) 
# 4.  75% data 
# 5.  upper fence 
```

```python
sns.boxplot(df['Age'])
```

```python
sns.boxplot(df['Fare'])
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

# Bivariate Analysis 

```python
# 1. scatterplot(Numerical column - Numerical column) 
```

```python
sns.scatterplot(x = tips['total_bill'] , 
                y = tips['tip'])
```

```python
sns.scatterplot(tips['total_bill'] , tips['tip'] , hue = tips['sex'])
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips , hue = tips['sex'])
```

```python
sns.scatterplot(x="total_bill" ,
                y = "tip" , data = tips , 
                hue = tips['sex'] , style = tips['smoker'])
```

```python
# Barplot (Numerical - Categorical) 
```

```python
sns.barplot(x = df['Pclass'] , y = df['Age'])
```

```python
sns.barplot(df['Pclass'] , df['Age'] , hue = df['Sex']) 
```

```python
# Boxplot(Numerical-Categorical) 
```

```python
sns.boxplot(df['Sex'] , df['Age'] , hue = df['Survived'])
```

```python
# Distplot(Numerical-Categorical)

sns.distplot(df['Age'])
```

```python
sns.distplot(df[df['Survived']==0]['Age'] , hist = False)
```

```python
# HeatMap(Categorical-Categorical) 
p = pd.crosstab(df['Pclass'] , df['Survived'])
p
```

```python
sns.heatmap(p)
```

```python
((df.groupby('Pclass').mean()['Survived'])*100).plot(kind = 'bar')

# df.groupby('Pclass').mean()
```
